IMPLEMENTATION OF IRIS DATASET


```python
import pandas as pd
```

```python
file=pd.read_csv("/content/archive(9).zip")
print(file)
```

```python

from sklearn.linear_model import Perceptron
x=file[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y=file['Species']
```

```python
model=Perceptron()
model=model.fit(x,y)
```

```python
op=model.predict([[5.1,3.5,1.4,0.2]])
print(op)
```

Cross Validation

```python
from sklearn.model_selection import train_test_split
x=file[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y=file['Species']

X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import Perceptron
```

```python
model=Perceptron()
```

```python
model=model.fit(X_train,Y_train)
```

```python
op=model.predict(X_test)
```

```python
print(op)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,ConfusionMatrixDisplay
```

```python
cm=confusion_matrix(Y_test,op)
print(cm)
```

```python
print(classification_report(Y_test,op,labels=model.classes_))
```

```python
import matplotlib.pyplot as plt
```

```python
disp=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=model.classes_)
disp.plot()
plt.show()
```
